```python
#import the packages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings(action='ignore')
```

```python
#read the data 
df=pd.read_csv(r'C:\Users\Admin\Desktop\Recruitment_Data.csv')
```

```python
#check first five rows and columns using head() function
df.head()
```

```python
#check the shape of the dataframe
df.shape
```

```python
#check for the null values(missing values)
df.isnull().sum()
```

```python
#replacing NaN values with 'Other'
df['recruiting_source']=df['recruiting_source'].replace(np.NaN,'Other')
```

```python
#againvcheck for the null values(missing values)
df.isnull().sum()
```

```python
#chcek the data type 
df.dtypes
```

```python
df['recruiting_source'].value_counts()
```

```python
#Grouping and perform count on each group
df1=df.groupby('recruiting_source')['recruiting_source'].count()
print(df1)
```

```python
#Average Sales number Group by Recruiting Source
sales_num = df.groupby('recruiting_source')['sales_quota_pct'].mean()
print(sales_num)
```

**1. Employees recruited from the Applied online and Other Source have Great performance than others**

```python
#average Attrition Number grouped by Recruiting Source
attr_num = df.groupby('recruiting_source')['attrition'].mean()
print(attr_num)
```

**1.Employees from the source other has less attrition numbers**

```python
sns.barplot(x=df['recruiting_source'],y=df['sales_quota_pct'],ci=0)
plt.title('Recruiting Sources VS Sales quota pct')
plt.show()
```

**1. Recruiting source Other has a higher performance**

**2. Sources like Applied online and Referral have nearly same performance**

```python
sns.barplot(x=df['recruiting_source'],y=df['attrition'],ci=0)
plt.title('Recruiting Sources VS Attrition')
plt.show()
```

**1. Recruiting source Other has a less attrition numbers**

```python
sns.barplot(x=df['recruiting_source'],y=df['performance_rating'],ci=0)
plt.title('Recruiting Sources VS Performance rating')
plt.show()
```

**1.Recruiting source like Applied online, Campus have same Performance Rating**

**2.Recruiting source Seach firm has less perfomance rating than all other sources**
